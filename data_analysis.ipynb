```python
# imports
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn import *
```

```python
# Load CSV into a pandas dataframe, we get an error because a lot of columns that should be ints also have 
# "NaN" strings thrown in. We will need to remove these
df = pd.read_csv("data/college_data.csv", na_values=["NaN", "PrivacySuppressed", "nan", "NAN", "none", "None", "NONE"])
```

```python
# Get metadata about the dataframe
df.info()
```

```python
# Look at the first few elements in our dataframe
df.head()
```

```python
# List the column headers of the data, this lets us figure out which data columnms we want to keep
list(df.columns)
```

```python
# Let's extract only the columns we feel are relevant
df_pruned = df[[
    "ACCREDAGENCY", # the accreditation agency will probably play into how "good" a uni is
    "SAT_AVG_ALL",
    "ACTCMMID",
    "ACTMTMID",
    "ACTWRMID",
    "AVGFACSAL",
    "PCTPELL", # percentage of students that are on the Pell grant
    "ENRL_4YR_TRANS_YR2_RT", # 4-year retention rate
    "ACTENMID",
    "GRAD_DEBT_MDN10YR_SUPP",
    "HI_INC_ENRL_ORIG_YR4_RT", # hi income enrollment
    "ADM_RATE", # our y-value
]].copy()
```

```python
df_pruned.info()
df_pruned.head()
```

```python
# Encode ACCREDAGENCY as a one-hot vector
df_pruned["ACCREDAGENCY"] = pd.DataFrame(df_pruned["ACCREDAGENCY"], dtype=str) # convert objects to strings
encoder = sk.preprocessing.LabelBinarizer()
agency_encoded = encoder.fit_transform(df_pruned["ACCREDAGENCY"])
agency_encoded # one-hot vector
```

We can't process strings, we need to have numeric data. We are going to replace the categorical data with sparse vectors of 1's. Then, each vector represents some category, and the vector will have a 1 at the index of the data it applies to. For example, if we have some data

    'data'
    a        
    b
    c
    
We can represent this as

    vec[a]    vec[b]   vec[c]
    1         0        0
    0         1        0
    0         0        1

```python
# Drop the old accreditation metrics and add our one hot vectors from before

agency_enc_df = pd.DataFrame(agency_encoded)
df_pruned.join(agency_enc_df)
del df_pruned["ACCREDAGENCY"]
```

```python
df_pruned.info()
df_pruned.head()
```

```python
# drop any rows where we don't have an ADM_RATE (that is our prediction value)
df_pruned = df_pruned[pd.notnull(df['ADM_RATE'])]
df_pruned.info()
```

Here we are imputing missing values. Our model is going to transform our data into matrices, and it can't deal with not having a value in a cell, or worse, having a string in a cell. Thus, we must either remove our cells without proper numeric data, or we can fill them in with 0's, the median, or the mean. In this case, we're going to fill our missing data in with the mean.

```python
# Fill in missing values for all float values with the average for each column
imputer = sk.preprocessing.Imputer(strategy='mean', axis=1)
df_imputed = pd.DataFrame(imputer.fit_transform(df_pruned))

# Restore original columns and indices
df_imputed.columns = df_pruned.columns
df_imputed.index = df_pruned.index

# normalize data
min_max_scaler = preprocessing.MinMaxScaler()
df_scaled = pd.DataFrame(min_max_scaler.fit_transform(df_imputed))
df_scaled.columns = df_pruned.columns
df_scaled.index = df_pruned.index
df_scaled.info()  # you can see that all the float columns have 2171 values
```

Now we're going to create a correlation matrix, and see which attributes correlate most highly with our y-value (the factor we're trying to predict) -- `ADMIT RATE`

```python
corr_matrix = df_imputed.corr()

corr_matrix["ADM_RATE"].sort_values(ascending=True)
```

We're trying to find the most selective schools, so it seems like the biggest factors that correlate to this are:
- average faculty salary
- SAT score
- high-income enrollment

Interestingly, the percentage of pell students do not correlate with this, and neither do individual ACT scores.

Now we have to pick a model and evaluate it. Let's start out with a Linear Regression using cross validation. With cross validation, we partition the dataset into `n` parts, and we train on `n-1` parts, saving the `nth` part for evaluation, and we run this `n-1` times, so our final error rate is an average of the error on all of the data (albeit computed with slightly models that are operating on different partitions of the data for each iteration). This makes our model more robust and helps prevent overfitting.
